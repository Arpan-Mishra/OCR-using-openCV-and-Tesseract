```python
import pytesseract
import cv2
import matplotlib.pyplot as plt
pytesseract.pytesseract.tesseract_cmd = r'C:\Users\Arpan\AppData\Local\Tesseract-OCR\tesseract.exe'
```

```python
img = cv2.imread('sample_img.jpeg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,8))
plt.imshow(img)
plt.axis('off')
plt.show()
```

Pre-process Image

```python
# all pixel intensities are set to 0 for intensity<120
ret,thresh = cv2.threshold(img,120,255,cv2.THRESH_TOZERO) 
# this makes the text pop out 

# plotting the processed image
plt.figure(figsize=(10,8))
plt.axis('off')
plt.imshow(thresh)
plt.show()
```

```python
# converting the image to string using pytesseract: https://pypi.org/project/pytesseract/

text = pytesseract.image_to_string(thresh,lang='eng')
print(text)

# saving the text
file = open('medicine-text.txt','w')
file.write(text)
file.close()
```
